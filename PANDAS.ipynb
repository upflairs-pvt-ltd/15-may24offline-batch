```python
# df 
# quantity , price  = total_price (quantity * price ) 
# python  
# df['bike_age'].min()  
# df 
# price less than 50k normal or middle class bike 
# price greater than 50K and  less than 1 lakh  --> high level bikes 
# 1 lakhs highest == > super bike 

# python  
ls = []
for price in df['price']: 
    if price <=50000: 
        ls.append('middle class bike')
    elif price > 50000 and price < 100000: 
        ls.append('high class bike')
    elif price> 100000: 
        ls.append("super bike")
    else:
        ls.append('i dont know')

# rename , cut()
        
```

```python
df['bike_class'] = ls 
```

```python
df[['price','bike_class']]
```
