```python
population_india = [100,200,222,300,400,800]
population_china = [80,180,200,250,300,700]
population_japan = [650,750,850,1000,1500,1560]

year =  [14,15,16,17,18,19]
plt.plot(year,population_india,marker="*" ,color='red',ms=8,mfc='yellow',lw=3,ls="dashed",label='india') # dotted 
plt.plot(year,population_china,marker="*" ,color='blue',ms=8,mfc='yellow',lw=3,ls="dotted",label='china') # dotted 
plt.plot(year,population_japan,marker="*" ,color='black',ms=8 ,mfc='yellow',lw=3,label='japan') # dotted 

plt.legend()
plt.title("population trend")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show() 
```

### seaborn 

```python
import seaborn as sns 
```

```python
# bike ,  ## pdfreader --> os, time, random, numpy 
tip = sns.load_dataset('tips')
tip.head()
```

```python
tip.columns = ['total_bill', 'tip', 'gender', 'smoker', 'day', 'time', 'size']
```

```python
tip.head()  # tips 
```

```python
# sns.scatterplot() 
# sns.boxenplot()  
# sns.barplot() 
# sns.countplot()
# sns.lineplot() 
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tip)
plt.show()  
```

```python
# relplot()   --> relational plot  (numerical,numerical)
# catplot()   --> categorical plot (numerical , categorical) 
# using only two functions, you can plot any chart 
```

```python
tip 
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='gender') 
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='gender',style='smoker') 
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='gender',style='smoker',col='gender') 
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='gender',style='smoker',col='day') 
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='time',style='smoker',col='day') 
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='time',style='smoker',col='day',kind='line') 
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,kind='line') 
plt.show()
# by default 'scatterplot' 
```

```python
# relplot()   --> scatter plot , lineplot 
# catplot()   --> (numerical ,categorical) 
# 
# univariet , bivariet , multivariet analysis  
# univariet analysis --> single column or single variable 
# bivariet -->   two columns or two variables  
# multivariet analysis --> multiple variables  
```
